```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
data = pd.read_csv("sales.csv")
data.head(5)

del data["Channel"]
del data["Region"]

data.head(5)
```

```python
custo = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=1)
    kmeans.fit(data)
    custo.append((i, kmeans.inertia_))
    
plt.xlabel("Numero de clusters")
plt.ylabel("Custo")
plt.plot([w[0] for w in custo], [w[1] for w in custo], marker="X")
```

```python
data_array = data.values

kmeans = KMeans(n_clusters=5, n_init=10, random_state=1234)
data["clusters"] = kmeans.fit_predict(data_array)

data.groupby("clusters").aggregate("mean").plot.bar(figsize=(10,7.5))
plt.title("Gastos por cluster")
```

```python
data["clusters"].value_counts()
```

```python
import numpy as np
import pandas as pd
import seaborn

from skimage.io import imread
from skimage.transform import resize
from skimage.color import rgb2hsv, hsv2rgb

from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm
from matplotlib.patches import Patch

seaborn.set()
%matplotlib inline
```

```python
current_image = imread("maurilio.jpg")
current_image = resize(current_image, (256,256), mode="reflect") #, anti_alising=True)
current_image_hsv = rgb2hsv(current_image)
```

```python
pd.DataFrame(current_image_hsv.reshape(-1,3), columns = ["R", "G", "B"]).head()
```

```python
pixel_list = current_image_hsv.reshape(-1,3)
```

```python
inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=1234)
    kmeans.fit(pixel_list)
    inertia.append((i,kmeans.inertia_))    
plt.plot([w[0] for w in inertia], [w[1] for w in inertia], marker="X")
plt.xlabel("Numero de Clusters")
plt.ylabel("Inercia (Custo)")
```

```python
kmeans = KMeans(n_clusters=3, n_init=10, random_state=1234)
kmeans.fit(pixel_list)
cluster_list = kmeans.labels_
cluster_mask = cluster_list.reshape(256,256)
```

```python
cmap = ListedColormap(hsv2rgb(kmeans.cluster_centers_.reshape(1,3,3)).reshape(3,3))
bounds=[0,1,2]
norm = BoundaryNorm(bounds, cmap.N)
plt.imshow(cluster_mask, cmap=cmap)
patches = [Patch(color=c, label="Cluster {}".format(i)) for i,c in enumerate(cmap.colors)]
plt.legend(handles=patches, loc=4, frameon=True)
```

```python
masked_image = np.dstack((current_image, cluster_mask==1))
plt.imshow(masked_image)
plt.grid(False)
```

```python
masked_image = np.dstack((current_image, cluster_mask!=1))
plt.imshow(masked_image)
plt.grid(False)
```
