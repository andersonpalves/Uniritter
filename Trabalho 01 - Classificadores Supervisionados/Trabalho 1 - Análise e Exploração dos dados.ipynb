```python
%matplotlib notebook
from matplotlib import style
import numpy as np
import pandas as pd

dataset = pd.read_csv('datasets/spotify.csv')
```

# 1. acoustiness X danceability

```python
style.use("seaborn-colorblind")
dataset.plot(x='acousticness', y='danceability', c='target', kind='scatter', colormap='Accent_r')
```

# 2. tempo X speechiness

```python
style.use("seaborn-colorblind")
dataset.plot(x='tempo', y='speechiness', c='target', kind='scatter', colormap='Accent')
```

# 3. danceability X energy

```python
style.use("seaborn-colorblind")
dataset.plot(x='danceability', y='energy', c='target', kind='scatter', colormap='winter')
```
